```python
outliers_iqr = {}
for column in train_data.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = train_data[column].quantile(0.25)
    Q3 = train_data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers_iqr[column] = train_data[(train_data[column] < (Q1 - 1.5 * IQR)) | (train_data[column] > (Q3 + 1.5 * IQR))].shape[0]

print(outliers_iqr)
print(max(outliers_iqr.items(), key=lambda x: x[1]))
```
